```python
# Imports 
import csv
from datetime import date as dt
import os

FILE_NAME = "expense.csv"
```

```python
# defining the function for adding expenses

def add_expense(category, amount, description =""):
    today = dt.today().isoformat()

    expense = {
        "date": today, 
        "category": category,
        "amount": amount,
        "description": description

    }

    headers = ["Date", "Category", "Amount", "Description"]

    
    with open(FILE_NAME, mode = "a", newline="") as file:
        writer = csv.writer(file)   

        if os.path.getsize(FILE_NAME) == 0 or os.path.exists(FILE_NAME):
            writer.writerow(headers)

        writer.writerow([today, category, amount, description])

add_expense("Niggas", 0.99, "Slavery")
```

```python
def read_expense():
    expense_list = []

    with open(FILE_NAME, mode = "r", newline="") as f:
        reader = csv.reader(f)
        next(reader) #Skips first row of headers 

        for row in reader:

            expense_list.append({"date": row[0],
                                 "category": row[1],
                                 "Amount": float(row[2]),
                                 "Description": row[3]
            
            })

    print(expense_list)
```
